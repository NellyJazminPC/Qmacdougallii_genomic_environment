```python
#Importar librerias
import ipyrad as ip
```

```python
## Crear un objeto Ensamble llamado data1. 
data3 = ip.Assembly("data3")
```

```python
## setting/modifying parameters for this Assembly object
data3.set_params('project_dir', "../../results/ipyrad_qlobata_qmacdou_de_novo/")
data3.set_params('sorted_fastq_path', "../../data/0.2_Trimmomatic/*.fastq.gz")
data3.set_params('assembly_method', "denovo")
data3.set_params('datatype', 'ddrad')
data3.set_params('output_formats', ['p', 's', 'k', 'g', 'v'])

## prints the parameters to the screen
data3.get_params()
```

Paso 1: 

```python
## run step 1 to create Samples objects
data3.run("1", auto=True, force=True)
```
