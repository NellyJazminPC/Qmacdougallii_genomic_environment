```python
## Crear un objeto Ensamble llamado data1. 
denovo_8 = ip.Assembly("denovo_8")
```

```python
## prints the parameters to the screen
denovo_8.get_params()
```

```python
## setting/modifying parameters for this Assembly object
denovo_8.set_params('sorted_fastq_path', "../ref_map_prueba/input/clean/*.fastq")
denovo_8.set_params('min_samples_locus', 2)
denovo_8.set_params('datatype', 'ddrad')
denovo_8.set_params('output_formats', ['p', 's', 'k', 'g', 'v'])

## prints the parameters to the screen
denovo_8.get_params()
```
